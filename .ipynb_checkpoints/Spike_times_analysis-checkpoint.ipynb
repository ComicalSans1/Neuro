```python
all_intervals = []

def restrict_spike_intervals(spike_times, s_min, s_max):
    for s_t in spike_times:
        restricted_intervals = [n for n in s_t if n >= s_min and n <= s_max]
        all_intervals.append(restricted_intervals)
    return all_intervals
    
restricted_spike_intervals = restrict_spike_intervals(spike_times, 5, 15)
```

```python
all_restricted_intervals = np.concatenate(restricted_spike_intervals)

print("Length of all intervals:", len(np.concatenate(spike_times)))
print("Length of restricted:", len(all_restricted_intervals))
print(f"Mean interval: {np.mean(all_restricted_intervals):.2f}")
print(f"Median interval: {np.median(all_restricted_intervals):.2f}")
print(f"Maximum interval: {np.max(all_restricted_intervals):.2f}")
```

We attempt to simulate this spiking behaviour using a linear integrate-and-fire neuron. This has a purely excitatory poisson input, and voltage resets to 0 after spiking with no refractory period. This admittedly simple model still satisfies the basics of spiking based on input.

```python
def lif_neuron_naive(n_steps=1000, alpha=0.02, rate=10):
    exc = stats.poisson(rate).rvs(n_steps)
    v = np.zeros(n_steps)
    spike_times = []

    for i in range(1, n_steps):
        dv = alpha * exc[i]
        v[i] = v[i-1] + dv
      
        if v[i] > 1:
            spike_times.append(i)
            v[i] = 0

    return v, spike_times

v, spike_times = lif_neuron_naive()
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 4))

ax1.plot(v[0:100])
for x in spike_times:
    if x <= 100:
        ax1.axvline(x, color="red")
ax2.hist([t - s for s, t in zip(spike_times[:-1], spike_times[1:])], np.linspace(0, 20));
```

In the case of constant excitatory input, we observe spikes at regular intervals, with the ISI histogram looking Gaussian with values evenly around a mean value. Evidently, this is not reflective of real-world spike data, which follows a an exponential distribution. This implies that the mechanism to generate spikes is fundamentally different from a simple excitatory input.  
Increasing alpha here increases the frequency of spikes, which makes sense as higher currents(alpha is a scalar multiplier to current) leads to higher excitation of the neuron.  

Biological neurons recieve both excitatory and inhibitory input, and also passively leak charge across the membrane. To simulate this, we sample from separate excitatory and inhibitory distributions.

```python
import scipy
import scipy.stats as stats

def lif_neuron(has_inh=True, n_steps=1000, alpha=0.5, beta=0.1, a_rate=10, b_rate=10):
    """ Simulates a leaky integrate-and-fire model of the neuron.

        the equation: dv = alpha * (exc[i] - inh[i]) - beta * v[i-1]
        "exc" and "inh" respectively, represent excitatory and inhibitory signals fired by presynaptic neurons. 
        These signals are simulated here using a Poisson distribution.
        To simulate passive leakage of charge across the membrane, we also have a leak current represented by leakage rate beta times the voltage of the membrane.
        
        The spiking frequency follows an exponential distribution because the firing probability of a neuron in a time period dt 
        is independent of previous time intervals, and so satisfies the memoryless property needed for exponential distributions?
    """
    exc = stats.poisson(a_rate).rvs(n_steps)
    inh = stats.poisson(b_rate).rvs(n_steps) * int(has_inh)

    v = np.zeros(n_steps)
    spike_times = []

    for i in range(1, n_steps):
        dv = alpha * (exc[i] - inh[i]) - beta * v[i-1]
        v[i] = v[i-1] + dv
      
        if v[i] > 1:
            spike_times.append(i)
            v[i] = 0

    return v, spike_times

np.random.seed(12)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(16, 4))

v, spike_times = lif_neuron()

ax1.plot(v[0:200])
for x in spike_times:
    if x <= 200:
        ax1.axvline(x, color="red")

ax2.hist([t - s for s, t in zip(spike_times[:-1], spike_times[1:])], np.linspace(0, 20))
ax2.set_xlim(0, 40)

plt.show()
```

The ISI histogram we observe here matches our observed data much more closely, with the ISIs following an exponential distribution.  

Now that we have modelled the behaviour of neurons, we will attempt to figure out why spike timings follow this specific distribution.  
Regarding biological neurons, one objective regarding interspike intervals is to maximize the information transmitted by each spike. One way to formalize this notion of information is the Shannon entropy, which is defined as $$-\sum p(x_i) * log_{2}p(x_i)$$

```python
def entropy(pmf):
    total_entropy = 0
    for p in pmf:
        if p != 0:
            total_entropy -= p * np.log2(p)
    return total_entropy

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(12, 3))

n_bins = 50  # number of points supporting the distribution
bins = np.linspace(0, 1, n_bins )  # bin edges

pmf = np.zeros(n_bins)
pmf[len(pmf) // 2] = 1.0  # middle point has all the mass

pmf_1 = np.zeros(n_bins)
pmf_1[len(pmf_1) // 3] = pmf_1[2 * len(pmf_1) // 3] = 0.5

pmf_2 = np.array([1/n_bins] * n_bins)

ax1.plot(bins, pmf, drawstyle="steps")
ax1.fill_between(bins, pmf, step="pre", alpha=0.4)

ax2.plot(bins, pmf_1, drawstyle="steps")
ax2.fill_between(bins, pmf_1, step="pre", alpha=0.4)

ax3.plot(bins, pmf_2)
ax3.fill_between(bins, pmf_2, step="pre", alpha=0.4)

plt.xlim(0, 1)
plt.ylim(0, 1)
plt.show()
```

```python
print(entropy(pmf)) # 0 entropy, as only 1 value is possible.
print(entropy(pmf_1)) # -(0.5log2(0.5) + 0.5log2(0.5)) = 1
print(f"{entropy(pmf_2):.2f}") # entropy of a uniform distribution over n values is simply log2(n)
```

Here we find that the maximally entropic distribution is a uniform distribution. This is inconvenient in the real world because this would mean that for large intervals the maximally entropic is just infenitesimally small probs across an infinite distance. Thus, the implied constraint is that each spike costs some energy to produce.  
The objective can then be reframed as maximixation of entropy **given a fixed number of spikes.**
